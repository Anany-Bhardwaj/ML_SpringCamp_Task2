```python
import energyflow
X,y=energyflow.qg_jets.load(num_data=100000, pad=True, ncol=4, generator='pythia',
                        with_bc=False, cache_dir='~/.energyflow')
```

```python
# Flatten the features for each jet
num_particles = X.shape[1]
num_features = X.shape[2]

# Create column names for each particle and feature
columns = [f"particle_{i}_{feature}" for i in range(num_particles) for feature in ['pt', 'y', 'phi', 'pid']]

# Reshape X to a 2D array
X_flattened = X.reshape(X.shape[0], -1)  # Shape: (num_data, num_particles * num_features)

# Convert to DataFrame
df_X= pd.DataFrame(X_flattened, columns=columns)
df_X['Labels']=y
```

```python
df = pd.DataFrame({
    'std_pt': np.std(X[:, :, 0], axis=1),
    'std_y': np.std(X[:, :, 1], axis=1),
    'std_phi': np.std(X[:, :, 2], axis=1),
    'std_pid': np.std(X[:, :, 3], axis=1),
    'label': y  # Keep the labels unchanged
})
print(df.head())
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
#correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
X1 = df['std_pt']
Y1 = df['label']
X_train , X_test , y_train , y_test = train_test_split(X1,Y1,test_size=0.2,random_state=42)
lr = LogisticRegression()
Scaler = StandardScaler()
X_train_scaled = Scaler.fit_transform(X_train.values.reshape(-1, 1))
X_test_scaled = Scaler.transform(X_test.values.reshape(-1, 1))
lr.fit(X_train_scaled, y_train)
y_pred = lr.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:\n", classification_report(y_test, y_pred))
```
